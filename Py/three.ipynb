```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv("C:/Users/predator/Downloads/salary_data.csv")
df.info()
```

```python
mean_ser = df.mean()
```

```python
x_mean = mean_ser.iloc[0]
y_mean = mean_ser.iloc[1]
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
```

```python
numi = pd.Series((df['YearsExperience']-x_mean)*(df['Salary']-y_mean))
```

```python
num = numi.sum()
num
```

```python
deni = pd.Series((df['YearsExperience']-x_mean)**2)
```

```python
den = deni.sum()
den
```

```python
b1 = num/den
b1
```

```python
b0 = y_mean-(b1*x_mean)
b0
```

```python
x = df['YearsExperience']
y = b0 + (b1*x)
plt.scatter(df['YearsExperience'],df['Salary'])
plt.plot(x,y,'-r')
plt.title('Graph')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
```

```python
from sklearn.linear_model import LinearRegression
X = np.array(x)
X_reshaped = X.reshape(-1, 1)

model = LinearRegression()
model.fit(X_reshaped, y)

y_pred_sklearn = model.predict(X_reshaped)

plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred_sklearn, color="green", label="Regression Line (Scikit-Learn)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Simple Linear Regression using Scikit-Learn")
plt.legend()
plt.show()

print(f"Least Squares Method: Intercept (b0) = {b0}, Slope (b1) = {b1}")
print(f"Scikit-Learn: Intercept = {model.intercept_}, Coefficient = {model.coef_[0]}")
```
