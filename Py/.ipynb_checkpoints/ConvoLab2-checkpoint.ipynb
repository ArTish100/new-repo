```python
#Importing necessary libraries
import requests
#!pip install bs4
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_link():
    df = pd.DataFrame(columns = ['head_title','head_link'])
    url = "https://timesofindia.indiatimes.com/home/headlines"
    r = requests.get(url)
    soup = BeautifulSoup(r.content,'html.parser')
    for i in soup.find_all('span',attrs = {'class':"w_tle"}):
        head_link = i.find('a')
        # print(head_link)
        linklist = []
        if head_link is not None:
            linklist.append(head_link.get('title'))
        else:
            linklist.append('NaN')
        if head_link is not None:
            linklist.append("https://timesofindia.indiatimes.com"+head_link.get('href'))
        else:
            linklist.append('NaN')
        
        df.loc[len(df)] = linklist
    return df
```

```python
def get_linkdin(position, location):
  temp_url='https://www.linkedin.com/search/results/all/?keywords='+str(position).replace(" ", "%20")+'%20in%20'+str(location).replace(" ", "%20")+'&origin=GLOBAL_SEARCH_HEADER'
  url='https://in.linkedin.com/jobs/search?keywords='+str(position)+'&location='+str(location)+'&geoId=&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0&original_referer='
  print(url)
  print(temp_url)
  df = pd.DataFrame(columns=['Job_Title', 'Company_name', 'Job_link'])
  response = requests.get(url)

  soup = BeautifulSoup(response.content, 'html.parser')

  for d in soup.find_all('div',attrs={'class':"base-search-card__info"}):
    job_data = []
    job_title = d.find('h3',attrs={'class':"base-search-card__title"})
    company_name = d.find('h4',attrs={'class':'base-search-card__subtitle'})

    par = d.find_parent()

    job_link = par.find('a',attrs={'class':"base-card__full-link absolute top-0 right-0 bottom-0 left-0 p-0 z-[2]"})

    if job_title is not None:
      job_data.append(job_title.text.strip())
    else:
      job_data.append('Nan')

    if company_name is not None:
      job_data.append(company_name.text.strip())
    else:
      job_data.append('Nan')

    if job_link is not None:
      job_data.append(job_link.get('href'))
    else:
      job_data.append('Nan')

    df.loc[len(df)] = job_data

  return df
```
