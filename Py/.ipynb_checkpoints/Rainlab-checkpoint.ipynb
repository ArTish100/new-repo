```python
from nltk.corpus import PlaintextCorpusReader
path = "C:/Users/predator/Desktop/trypy" #use some txt file path for custom corpus
dataset = PlaintextCorpusReader(path,'.*') # 2nd param type of files to be read; '.*' is for all type of files
dataset.fileids()
```

```python
corpus = []
filenames = dataset.fileids()
for i in range(len(filenames)):
    corpus.append(dataset.raw(fileids = filenames[i]))
#saving content in list
```

```python
#saving content in dataframe
import pandas as pd
df = pd.DataFrame(columns=['text'])
for i in range(len(filenames)):
    df.loc[i] = [dataset.raw(fileids = filenames[i])]
print(df)
```

```python
import os
filenames1 = os.listdir(path)
```

```python
corpus1 = []
for i in range(len(filenames1)):
    f = open(path+'/'+filenames1[i])
    corpus1.append(f.read())
    f.close()
```

```python
no_stop = [] # no stopwords i.e. stopword removal
stopword_list = stopwords.words('english')
stopword_list
```

```python
for doc in norm_token:
    no_stop.append([word for word in doc if word not in stopword_list])

print(no_stop)
```

```python
from nltk.stem import PorterStemmer
final = []
ps = PorterStemmer()
for doc in no_stop:
    final.append([ps.stem(word) for word in doc])
```
